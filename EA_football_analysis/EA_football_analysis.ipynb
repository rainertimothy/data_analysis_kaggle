```python
df.drop('Unnamed: 0', axis=1, inplace=True)
```

OVR Rating


```python
# Top 10 Players with Highest Rating
ovr_rank = df.sort_values('OVR', ascending=False)
overall_rank = ovr_rank[['Rank', 'Name', 'Age', 'Nation', 'Team', 'OVR', 'PAC', 'SHO', 'PAS', 'DRI', 'DEF', 'PHY']].head(10)
overall_rank
```

```python
# Top 10 players visualization with barplot

plt.figure(figsize=(10, 6))
plt_bar = sns.barplot(x='OVR', y='Name', data=overall_rank, palette='flare', hue='Name')
for bar in plt_bar.patches:
    bar.set_height(0.8)
plt.title('Top 10 Players by Overall Rating')
for index, value in enumerate(overall_rank['OVR']):
    plt.text(value + 0.5, index, str(value), ha='left', va='center')
plt.xlabel('OVR')
plt.ylabel('Name')
plt.show()
```

```python
# Top 10 players and their attributes

player_attributes = overall_rank[['Name', 'PAC', 'SHO', 'PAS', 'DRI', 'DEF', 'PHY']]
melted = pd.melt(player_attributes, id_vars=['Name'], var_name='Attribute', value_name='Value')

plt.figure(figsize=(10, 10))
barplt = sns.barplot(x='Value', y='Name', hue='Attribute', data=melted, palette='flare')
for bar in barplt.patches:
    bar.set_height(0.118)
plt.title('Player Attributes')
plt.xlabel('Value')
plt.ylabel('Name')
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.show()
```

Player Attributes


```python
# correlation between player attributes

plt.figure(figsize=(10, 8))
heatmap_attributes = df[['OVR', 'PAC', 'SHO', 'PAS', 'DRI', 'DEF', 'PHY']].corr()
mask = np.triu(np.ones_like(heatmap_attributes, dtype=bool))
sns.heatmap(heatmap_attributes, mask=mask, annot=True, cmap='viridis', linewidths=.2)
plt.gca().set_facecolor('lightgrey')
plt.gca().patch.set_alpha(0.5)
plt.title('Correlation Heatmap of Player Attributes')
plt.show()
```

```python
# Player attributes impact on overall rating

attributes = ['PAC', 'SHO', 'PAS', 'DRI', 'DEF', 'PHY']

figure, ax = plt.subplots(2, 3, figsize=(10, 6))
figure.suptitle('Player Attributes Impact on Overall Rating')
for i, attribute in enumerate(attributes):
    sns.regplot(x='OVR', y=attribute, data=df, ax=ax[i//3, i%3], marker = '.', color='blue', line_kws=dict(alpha=0.8, color='red'))
    ax[i//3, i%3].set_title(f'{attribute} vs OVR')
    ax[i//3, i%3].set_xlabel('OVR')
    ax[i//3, i%3].set_ylabel(attribute)
plt.tight_layout()
plt.show()
```

```python
# distribution of player attributes

figure, ax = plt.subplots(2, 3, figsize=(10, 6), sharey=True)
figure.suptitle('Attributes Number Distribution')
for i, attribute in enumerate(attributes):
    sns.histplot(df[attribute], bins=20, ax=ax[i//3, i%3], kde=True)
    ax[i//3, i%3].set_xlabel(attribute)
plt.tight_layout()
plt.show()
```

Playing Styles


```python
# Top 10 playing styles

playing_style = df['play style'].str.split(', ')
playing_style = playing_style.explode()
playing_style = playing_style.value_counts()
top_styles = playing_style.head(10)

plt.figure(figsize=(10, 6))
sns.barplot(x=top_styles.values, y=top_styles.index, palette='viridis', hue=top_styles.index)
plt.title('Top 10 players style of play')
for x,y in enumerate(top_styles.values):
    plt.text(y, x, str(y), ha='right', color='white', va='center')
plt.xlabel('Number of players')
plt.ylabel('Playing style')
plt.show()
```

Player Age


```python
# Player age distribution

plt.figure(figsize=(10, 6))
sns.histplot(df['Age'], bins=20, color='darkblue', kde=True)
plt.grid(axis='y', alpha=0.2)
plt.title('Age Distribution')
```

```python
# Age impact on overall rating

plt.figure(figsize=(10, 6))
sns.regplot(x='OVR', y='Age', data=df, color='blue', marker='.', line_kws=dict(alpha=0.8, color='red'), x_jitter=0.5,
            y_jitter=0.5)
plt.title('Age vs Overall Rating')
plt.grid(True, linestyle='-', alpha=0.6)
plt.xlabel('OVR')
plt.ylabel('Age')
plt.show()
```

League, Nation, Clubs


```python
# Top 10 Highest Rated Leagues

top_leagues = df.groupby('League')['OVR'].mean().sort_values(ascending=False).head(10)
plt.figure(figsize=(10,6))
sns.barplot(x=top_leagues.values, y=top_leagues.index, palette='RdYlBu', hue=top_leagues.index)
plt.title('Top 10 Highest Rated Leagues')
for x, y in enumerate(top_leagues.values):
    plt.text(y + 0.5, x, f'{y:.1f}', ha='left', color='black')
plt.xlabel('Average OVR')
plt.ylabel('League')
plt.show()
```

```python
# top 10 highest rated leagues and their highest rated players

highest_per_team = df.loc[df.groupby('League')['OVR'].idxmax()][['League', 'Name', 'OVR']]
highest_per_team.reset_index(drop=True, inplace=True)
highest_per_team.sort_values('OVR', ascending=False).head(10).reset_index(drop=True)
```

```python
# Top 10 highest rated teams

top_teams = df.groupby('Team')['OVR'].mean().sort_values(ascending=False).head(10)
plt.figure(figsize=(10,6))
sns.barplot(x=top_teams.values, y=top_teams.index, palette='RdYlBu', hue=top_teams.index)
plt.title('Top 10 Highest Rated Teams')
for x, y in enumerate(top_teams.values):
    plt.text(y + 0.5, x, f'{y:.1f}', ha='left', color='black', va='center')
plt.xlabel('Average OVR')
plt.ylabel('Teams')
plt.show()
```

```python
# top 10 highest rated teams and their highest rated players

highest_per_team = df.loc[df.groupby('Team')['OVR'].idxmax()][['Team', 'Name', 'OVR']]
highest_per_team.reset_index(drop=True, inplace=True)
highest_per_team.sort_values('OVR', ascending=False).head(10).reset_index(drop=True)
```

```python
# Top 10 Highest Rated Nations

top_nations = df.groupby('Nation')['OVR'].mean().sort_values(ascending=False).head(10)
plt.figure(figsize=(10,6))
sns.barplot(x=top_nations.values, y=top_nations.index, palette='RdYlBu', hue=top_nations.index)
plt.title('Top 10 Highest Rated Nations')
for x, y in enumerate(top_nations.values):
    plt.text(y + 0.5, x, f'{y:.1f}', ha='left', color='black')
plt.xlabel('Average OVR')
plt.ylabel('League')
plt.show()
```

```python
# Find the highest OVR player for each nation
highest_ovr_per_nation = df.loc[df.groupby('Nation')['OVR'].idxmax()][['Nation', 'Name', 'OVR']]

# Display the result
highest_ovr_per_nation.reset_index(drop=True, inplace=True)
highest_ovr_per_nation = pd.DataFrame(highest_ovr_per_nation)
highest_ovr_per_nation.sort_values('OVR', ascending=False).head(10).reset_index(drop=True)
```
